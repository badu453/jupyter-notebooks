```python
import numpy as np
import pandas as pd

data = pd.read_csv('file (4).csv')
data1=pd.read_csv('file (5).csv')
data['Gender'] = (
    data.groupby('Name')['Gender']
      .transform(lambda s: s.ffill().bfill().infer_objects(copy=False))
)

data1['Gender'] = (
    data1.groupby('Name')['Gender']
      .transform(lambda s: s.ffill().bfill().infer_objects(copy=False))
)
```

```python
data['Gender'] = data['Gender'].replace({'Female': 1, 'Male': 0})

data1['Gender'] = data1['Gender'].replace({'Female': 1, 'Male': 0})

data.loc[(data['Name'] == 'Sara') & (data['Gender'].isna()), 'Gender'] = 1
data.loc[((data['Name'] == 'Sajjad') | (data['Name'] == 'Qasim') ) & (data['Gender'].isna()), 'Gender'] = 0
data1.loc[(data1['Name'] == 'Sara') & (data1['Gender'].isna()), 'Gender'] = 1
data1.loc[((data1['Name'] == 'Sajjad') | (data1['Name'] == 'Qasim') ) & (data1['Gender'].isna()), 'Gender'] = 0



data.drop(['ID','Name'], axis=1, inplace=True)
data1.drop(['ID','Name'], axis=1, inplace=True)
```

```python
data['Age'] = data['Age'].replace({np.nan: 20, -1: 20})
data.dropna(subset=['Admission Status'], inplace=True)
data.dropna(subset=['Gender'], inplace=True)

data1['Age'] = data1['Age'].replace({np.nan: 20, -1: 20})
# data1.dropna(subset=['Admission Status'], inplace=True)
data1.dropna(subset=['Gender'], inplace=True)
data = data.drop(80, axis=0)
data = data.drop(49, axis=0)
data
```

```python
data = data.drop_duplicates()
data1 = data1.drop_duplicates()
data
```

```python
from sklearn.preprocessing import OneHotEncoder


data['City'] = data['City'].fillna('nan')

data1['City'] = data1['City'].fillna('nan')

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)


city_encoded = encoder.fit_transform(data[['City']])
city_encoded1 = encoder.fit_transform(data1[['City']])

city_df = pd.DataFrame(city_encoded, columns=encoder.get_feature_names_out(['City']), index=data.index)
city_df1 = pd.DataFrame(city_encoded1, columns=encoder.get_feature_names_out(['City']), index=data1.index)


data = pd.concat([data, city_df], axis=1)
data1 = pd.concat([data1, city_df1], axis=1)
data = data.drop('City', axis=1)
data1 = data1.drop('City', axis=1)





data['High School Percentage'] = data['High School Percentage'].replace({110.5: 75, -10.0: 75, np.nan: 75})
data['Admission Test Score'] = data['Admission Test Score'].replace({150.0: 77, np.nan: 75})
data1['High School Percentage'] = data['High School Percentage'].replace({110.5: 75, -10.0: 75, np.nan: 75})
data1['Admission Test Score'] = data['Admission Test Score'].replace({150.0: 77, np.nan: 75})

# data['City'] = data['City'].fillna('Unknown')

data
```

```python
X = data.drop('Admission Status', axis=1)
y = data['Admission Status'].map({'Accepted': 1, 'Rejected': 0})

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)

data1
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np # Ensure numpy is imported for np.nan

# Split the data into training and testing sets

model = GradientBoostingClassifier(
    # n_estimators=300,
    # learning_rate=0.03,
    # max_depth=4,
    # subsample=0.8,
    # random_state=2026
)


# Train the model
model.fit(X, y)

# Ensure data1 has no NaNs before prediction
# Applying the same replacement logic as intended for data1 in the preprocessing steps
data1['High School Percentage'] = data1['High School Percentage'].replace({110.5: 75, -10.0: 75, np.nan: 75})
data1['Admission Test Score'] = data1['Admission Test Score'].replace({150.0: 77, np.nan: 75})

# Make predictions on the test set
y_pred = model.predict(data1)
data1
```

```python
# Load the original IDs but filter them using the index of data1 which survived preprocessing
original_data1 = pd.read_csv('file (5).csv')
predicted_ids = original_data1.loc[data1.index, 'ID']

sub = pd.DataFrame({"subtaskID": 1, "datapointID": predicted_ids, "answer": y_pred})
sub.to_csv("submission.csv", index=False)
sub
```
